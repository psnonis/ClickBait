# Feature Engineering : Categorical Variables

## Replace Undefined Values with Special Term

```python
Engineering.catFillUndefined(subset = 'train', iStep = f'normed')
Engineering.catFillUndefined(subset = 'tests', iStep = f'normed')
Engineering.catFillUndefined(subset = 'valid', iStep = f'normed')
```

```python
Engineering.pdf(f'data/train.parquet.normed.filled', rows = 20)
```

## Find Uncommon Values for Each Feature

```python
Engineering.catFindFrequents(subset = 'train', iStep = f'normed.filled', min = 60000, fit = True)
```

## Replace Uncommon Values with Special Term

```python
Engineering.catMaskUncommons(subset = 'train', iStep = f'normed.filled', min = 60000)
Engineering.catMaskUncommons(subset = 'tests', iStep = f'normed.filled', min = 60000)
Engineering.catMaskUncommons(subset = 'valid', iStep = f'normed.filled', min = 60000)
```

## Encode Categorical Features

```python
Engineering.catDoCodeFeature(subset = 'train', iStep = f'normed.filled.masked-{60000:06d}', fit = True)
Engineering.catDoCodeFeature(subset = 'tests', iStep = f'normed.filled.masked-{60000:06d}')
Engineering.catDoCodeFeature(subset = 'valid', iStep = f'normed.filled.masked-{60000:06d}')
```

```python
Engineering.pdf(f'data/train.parquet.normed.filled.masked-{60000:06d}.encode', filter = 'label == 1', rows = 20)
```

```python
Engineering.pdf(f'data/train.parquet.normed.filled.masked-{60000:06d}.encode', filter = 'label == 0', rows = 20)
```

## Pick Categorical Features

```python
Engineering.catDoPickFeature(subset = 'train', iStep = f'normed.filled.masked-{60000:06d}.encode', top = 987, fit = True)
Engineering.catDoPickFeature(subset = 'tests', iStep = f'normed.filled.masked-{60000:06d}.encode', top = 987)
Engineering.catDoPickFeature(subset = 'valid', iStep = f'normed.filled.masked-{60000:06d}.encode', top = 987)
```

```python
Engineering.pdf(f'data/train.parquet.normed.filled.masked-{60000:06d}.encode.picked-{987:06d}', filter = 'label == 1', rows = 20)
```

# Combine All Features

```python
Engineering.allDoPackFeature(subset = 'train', iStep = f'normed.filled.masked-{60000:06d}.encode.picked-{987:06d}', fit = True)
Engineering.allDoPackFeature(subset = 'tests', iStep = f'normed.filled.masked-{60000:06d}.encode.picked-{987:06d}')
Engineering.allDoPackFeature(subset = 'valid', iStep = f'normed.filled.masked-{60000:06d}.encode.picked-{987:06d}')
```

# Take Toy Sample
