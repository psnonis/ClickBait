## Import Dependencies

```python
%load_ext autoreload
%autoreload 2

from code.engineering import *
```

## Setup Spark Application

```python
Engineering.setupSpark(application = 'prep')
```

# Load Criteo Dataset

```python
Engineering.importData(location = 'data', clean = True)
```

## Split Data into Training, Test, and Dev Subsets

```python
Engineering.splitsData(ratios = [.8, .1, .1])
```

# Initial Feature Engineering

```python
min = 60000 # 1150
top =   987 #  987 + 13 = 1000
```

## Calculate Numerical Feature Statistics

```python
Engineering.numDoMeasurement(subset = 'train', iStep = f'')
```

```python
Engineering.num_measures
```

## Standardize Numerical Features and Fill Undefined Categorical Features

```python
Engineering.allDoStandardize(subset = 'train', iStep = f'')
Engineering.allDoStandardize(subset = 'tests', iStep = f'')
Engineering.allDoStandardize(subset = 'valid', iStep = f'')
```

```python
Engineering.pdf(f'data/train.parquet.normed', rows = 20)
```

## Calculate Categorical Feature Statistics

```python
Engineering.catDoMeasurement(subset = 'train', iStep = f'normed')
```

```python
for feature, measure in Engineering.cat_measures.items():
    print(f'{feature} : features = {len(measure):>7} : > {min} = {sum(measure.counts > min)}')
```

## Replace Uncommon Categorical Values with Special Term

```python
Engineering.catMaskUncommons(subset = 'train', iStep = f'normed', min = 60000)
Engineering.catMaskUncommons(subset = 'tests', iStep = f'normed', min = 60000)
Engineering.catMaskUncommons(subset = 'valid', iStep = f'normed', min = 60000)
```

## Encode Categorical Features

```python
Engineering.catDoCodeFeature(subset = 'train', iStep = f'normed.masked-{60000:06d}', min = 60000, fit = True)
Engineering.catDoCodeFeature(subset = 'tests', iStep = f'normed.masked-{60000:06d}', min = 60000)
Engineering.catDoCodeFeature(subset = 'valid', iStep = f'normed.masked-{60000:06d}', min = 60000)
```

```python
Engineering.pdf(f'data/train.parquet.normed.masked-{60000:06d}.encode', filter = 'label == 1', rows = 20)
```

## Pick Categorical Features

```python
Engineering.catDoPickFeature(subset = 'train', iStep = f'normed.masked-{60000:06d}.encode', top = 987, fit = True)
Engineering.catDoPickFeature(subset = 'tests', iStep = f'normed.masked-{60000:06d}.encode', top = 987)
Engineering.catDoPickFeature(subset = 'valid', iStep = f'normed.masked-{60000:06d}.encode', top = 987)
```

```python
Engineering.pdf(f'data/train.parquet.normed.masked-{60000:06d}.encode.picked-{987:06d}', filter = 'label == 1', rows = 20)
```

# Combine All Features

```python
Engineering.allDoPackFeature(subset = 'train', iStep = f'normed.masked-{60000:06d}.encode.picked-{987:06d}', fit = True)
Engineering.allDoPackFeature(subset = 'tests', iStep = f'normed.masked-{60000:06d}.encode.picked-{987:06d}')
Engineering.allDoPackFeature(subset = 'valid', iStep = f'normed.masked-{60000:06d}.encode.picked-{987:06d}')
```

# Take Toy Sample

```python
Engineering.toyTakeSubSample(subset = 'train', iStep = f'normed.masked-{60000:06d}.encode.picked-{987:06d}.packed', len = 8000)
Engineering.toyTakeSubSample(subset = 'tests', iStep = f'normed.masked-{60000:06d}.encode.picked-{987:06d}.packed', len = 1000)
Engineering.toyTakeSubSample(subset = 'valid', iStep = f'normed.masked-{60000:06d}.encode.picked-{987:06d}.packed', len = 1000)
```
